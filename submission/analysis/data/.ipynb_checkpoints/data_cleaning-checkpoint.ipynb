```python
# Import the dependencies
import pandas as pd
import numpy as np
import datetime

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

# ORM Stuff
from sqlalchemy import create_engine, inspect, text, func
# from sqlalchemy.orm import Session
# from sqlalchemy.ext.automap import automap_base
```

```python
# Load the CSV file
file_path1 = "unclean-data/game_sales.csv"
file_path2 = "unclean-data/game_reviews.csv"
df_games = pd.read_csv(file_path1)
df_reviews = pd.read_csv(file_path2)
# Display the first few rows to understand its structure
df_games.head()
```

```python
platform_mapping = {
    " 3DS": "3DS",
    " DS": "DS",
    " Dreamcast": "DC",
    " Game Boy Advance": "GBA",
    " GameCube": "GC",
    " Nintendo 64": "N64",
    " PlayStation": "PS",
    " PlayStation 2": "PS2",
    " PlayStation 3": "PS3",
    " PlayStation 4": "PS4",
    " PlayStation Vita": "PSV",
    " Wii": "Wii",
    " Wii U": "WiiU",
    " Xbox": "XB",
    " Xbox 360": "X360",
    " Xbox One": "XOne",
}
# Apply the mapping
df_reviews["platform"] = df_reviews["platform"].map(platform_mapping)
df_reviews.head()
```

```python
# 2. Prepare df_reviews for merging
df_games.rename(columns={"Name": "name", "Platform": "platform"}, inplace=True)
```

```python
df_merged = pd.merge(
    df_games,
    df_reviews,
    on=["name", "platform"],
    how="left"
)
df_merged.info()
```

```python
df_merged = df_merged.rename(columns={"Year_of_Release": "Year", 
                            "NA_Sales": "NA",
                            "EU_Sales": "EU",
                            "JP_Sales": "JP",
                            "Other_Sales": "Other",
                            "Global_Sales": "Global"})
df_merged = df_merged[df_merged["Year"].notnull()]
df_merged = df_merged[df_merged["Genre"].notnull()]
df_merged = df_merged[df_merged["Publisher"].notnull()]
df_merged['Year'] = df_merged['Year'].astype(int)
df_merged["Critic_Score"] = df_merged["Critic_Score"].fillna(df_merged["meta_score"])
df_merged["User_Score"] = df_merged["User_Score"].fillna(df_merged["user_review"])
df_merged = df_merged.drop(columns=['User_Count', 'summary', 'user_review', 'meta_score', 'release_date'])
df_merged.info()
```

```python
df_merged.head()
```

```python
df_merged.to_csv("../data/game_sales_tableau_clean.csv", index = False)
```
